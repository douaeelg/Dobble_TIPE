## 2. Rappels théoriques

Dans un plan projectif fini d'ordre \(q\) :

- il y a \(q^2 + q + 1\) points ;
- il y a \(q^2 + q + 1\) droites ;
- chaque droite contient \(q + 1\) points ;
- deux droites distinctes se coupent en un unique point.


````python
afficher_figure(
    "tableau_parametres_plans_projectifs",
    tableau_parametres_svg(),
    "Tableau récapitulatif des tailles des plans projectifs finis.",
)
````
